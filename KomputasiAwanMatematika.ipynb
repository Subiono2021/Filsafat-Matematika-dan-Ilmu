Berikut adalah program visualisasi data menggunakan heatmap:

Hasil gambar heatmap adalah representasi visual dari matriks data numerik dalam bentuk grid, di mana setiap nilai pada matriks direpresentasikan oleh warna tertentu. Berikut penjelasan elemen-elemen penting pada gambar:

1. <b>Grid (Matriks 10x10)</b>:
   - Setiap sel dalam grid merepresentasikan satu elemen dalam matriks data numerik berukuran 10x10.
   - Nilai-nilai dalam matriks tersebut dihasilkan secara acak dan berkisar antara 0 hingga 1.

2. <b>Warna</b>:
   - Warna dalam heatmap diambil dari skema warna <b>viridis</b>, yang memiliki gradasi dari ungu gelap (nilai rendah) ke kuning terang (nilai tinggi).
   - Intensitas warna menunjukkan besar kecilnya nilai dalam matriks. Warna yang lebih terang biasanya berarti nilai lebih tinggi, sedangkan warna gelap menunjukkan nilai lebih rendah.

3. <b>Color Bar (Peta Warna)</b>:
   - Color bar di sisi kanan menunjukkan hubungan antara nilai numerik dan warna yang sesuai. Ini mempermudah interpretasi nilai berdasarkan warna yang terlihat pada heatmap.

4. <b>Sumbu X dan Y</b>:
   - Sumbu X dan Y menunjukkan indeks atau posisi elemen di dalam matriks.
   - Sebagai contoh, nilai di baris ke-2 dan kolom ke-3 pada matriks akan diwakili oleh sel yang sesuai di heatmap.

5. <b>Judul dan Label</b>:
   - Judul <em>"Sample Heatmap"</em> menggambarkan bahwa ini adalah contoh visualisasi data.
   - Label sumbu X dan Y menunjukkan bahwa nilai-nilai ini berada dalam dimensi matriks.

### Fungsi Heatmap:
Heatmap sering digunakan untuk memahami pola atau tren dalam data numerik besar, misalnya:
- Korelasi antar variabel.
- Distribusi intensitas dalam analisis spasial.
- Identifikasi area dengan nilai ekstrem (maksimum/minimum).

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(42)
data = np.random.rand(10, 10)  # 10x10 matrix with random values

# Create the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(data, cmap='viridis', aspect='auto')

# Add color bar and labels
plt.colorbar(label='Intensity')
plt.title('Sample Heatmap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

#  <b>4. AI dan Proses Pembelajaran Matematika</b>

<summary><button><b><mark>4.1. <em>Artificial Intelligence</em> (AI)</mark></b></button></summary>
<p align="justify">
<b>Kelahiran kecerdasan buatan (1941-1956):</b> Penelitian awal tentang mesin berpikir terinspirasi oleh pertemuan berbagai ide yang menjadi populer pada akhir 1930-an, 1940-an, dan awal 1950-an. Penelitian terbaru di bidang neurologi telah menunjukkan bahwa otak adalah jaringan listrik neuron yang melepaskan pulsa secara <em>all-or-nothing</em>. Cybernetics <b>Norbert Wiener</b> menggambarkan kontrol dan stabilitas dalam jaringan listrik. Teori informasi <b>Claude Shannon</b> menggambarkan sinyal digital (yaitu, sinyal <em>all-or-nothing</em>). Teori komputasi <b>Alan Turing</b> menunjukkan bahwa setiap bentuk komputasi dapat dijelaskan secara digital. Hubungan erat antara ide-ide ini menunjukkan bahwa mungkin untuk membangun "<b>otak elektronik</b>".
</p>
<p align="justify">
Pada tahun 1940-an dan 1950-an, segelintir ilmuwan dari berbagai bidang (matematika, psikologi, teknik, ekonomi, dan ilmu politik) mengeksplorasi beberapa arah penelitian yang akan sangat penting bagi penelitian AI selanjutnya. <b>Alan Turing</b> adalah salah satu orang pertama yang secara serius menyelidiki kemungkinan teoretis "kecerdasan mesin". Bidang "penelitian kecerdasan buatan" didirikan sebagai disiplin akademis pada tahun 1956.
</p>
<p align="justify">
<b>Tes Turing :</b>
Pada tahun 1950, Turing menerbitkan sebuah makalah penting berjudul "<em>Computing Machinery and Intelligence</em>", di mana ia berspekulasi tentang kemungkinan menciptakan mesin yang dapat berpikir. Dalam makalah tersebut, ia mencatat bahwa "berpikir" sulit didefinisikan dan merancang <b>Tes Turing</b> yang terkenal: Jika sebuah mesin dapat melakukan percakapan (melalui <b>teleprinter</b>) yang tidak dapat dibedakan dari percakapan dengan manusia, maka masuk akal untuk mengatakan bahwa mesin tersebut "<b>berpikir</b>". Versi sederhana dari masalah ini memungkinkan Turing untuk berargumen secara meyakinkan bahwa "<b>mesin berpikir</b>" setidaknya mungkin dan makalah tersebut menjawab semua keberatan paling umum terhadap proposisi tersebut. Tes Turing adalah proposal serius pertama dalam filsafat kecerdasan buatan.
</p>
<p align="justify">
Kecerdasan Buatan atau <em><b>Artificial Intelligence</b></em> (<b>AI</b>) adalah simulasi kecerdasan manusia dalam mesin yang diprogram untuk berpikir seperti manusia dan meniru tindakannya. AI dirancang untuk melakukan tugas-tugas yang biasanya membutuhkan kecerdasan manusia, seperti visualisasi, pengenalan ucapan, pengambilan keputusan, dan penerjemahan bahasa.

<summary><button><b><mark>4.4. Di Persimpangan dengan AI: Merangkul Resistensi dalam Pembelajaran dengan AI Generatif</mark></b></button></summary>
<p align="justify">
Perjalanan memprogram ulang cara kita belajar, khususnya dengan bantuan AI generatif, tak pelak lagi terkait erat dengan resistensi-resistensi yang berasal dari bias pribadi, pengondisian budaya, dan keyakinan yang mengakar kuat. Berikut ini adalah eksplorasi ide-ide tersebut secara mendetail:
</p>
<ol align="justify">
<li><b>Memahami Resistensi</b><br>
Resistensi terhadap AI, atau teknologi transformatif apa pun, sering kali muncul dari bias—kerangka kognitif yang dibentuk oleh pengalaman pribadi, sosial, dan budaya. Bias ini memiliki dua tujuan:
<ul align="justify">
<li><b>Mekanisme Perlindungan:</b> Bias melindungi kita dari ide-ide yang tidak dikenal atau berpotensi berbahaya.
</li>
<li><b>Penghalang Perubahan:</b> Bias membatasi kemampuan kita untuk beradaptasi, tumbuh, dan mengeksplorasi kemungkinan-kemungkinan baru. Misalnya:
<ul align="justify">
<li>Jika seseorang tumbuh dalam lingkungan yang skeptis terhadap teknologi, mereka mungkin secara inheren tidak mempercayai AI.
</li>
<li>Penggambaran negatif AI di media atau penolakan dari individu yang dipercaya (guru, rekan, atau pemimpin pemikiran) dapat memperkuat kekhawatiran.
</li>
</ul>
</li>
</ul>
</li>
<li><b>Bias sebagai Pedang Bermata Dua</b><br>
Bias pada dasarnya tidaklah buruk; bias membentuk cara kita memproses informasi dan membuat keputusan. Namun, dalam konteks AI generatif, bias dapat:
<ul align="justify">
<li><b>Memfasilitasi Pembelajaran:</b> Dengan menyediakan kerangka kerja yang familiar untuk pemahaman.
</li>
<li><b>Menghambat Pertumbuhan:</b> Dengan menghalangi eksplorasi potensi AI.
</li>
</ul>
Belajar ulang dengan AI generatif memerlukan penanganan bias ini secara cermat.
</li>
<li><b>Peran AI Generatif dalam Menantang Bias</b><br>
AI Generatif menantang paradigma pembelajaran tradisional, meminta kita untuk mempertimbangkan kembali pendekatan kita terhadap perolehan dan penerapan pengetahuan. Berikut caranya:
<ul align="justify">
<li><b>Mempertanyakan Norma yang Sudah Ada:</b> Wawasan yang dihasilkan AI dapat mendorong kita untuk memikirkan kembali keyakinan yang telah lama dipegang dengan menghadirkan perspektif alternatif.
</li>
<li><b>Memperluas Cakrawala:</b> Alat AI menawarkan akses ke berbagai sumber daya dan ide yang beragam yang melampaui batasan pribadi atau budaya.
</li>
<li><b>Menyesuaikan Pengalaman:</b> Dengan beradaptasi dengan kebutuhan individu, AI dapat menunjukkan nilai pendekatan pembelajaran yang inovatif dan disesuaikan.
</li>
</ul>
</li>
<li><b>Membuat Keputusan yang Tepat</b><br>
Setiap individu harus memutuskan apakah dan bagaimana cara menggunakan AI generatif. Ini melibatkan:
<ul align="justify">
<li><b>Mendidik Diri Sendiri:</b> Memahami teknologi-kemampuan, keterbatasan, dan implikasinya-adalah langkah pertama.
</li>
<li><b>Mengevaluasi Tujuan:</b> Jelaskan mengapa Anda mungkin ingin menggunakan AI. Apakah untuk meningkatkan efisiensi, mendapatkan perspektif baru, atau mendukung kreativitas?
</li>
<li><b>Mengakui Kekhawatiran:</b> Kekhawatiran etika, budaya, dan praktis tentang penggunaan AI adalah valid dan harus dipertimbangkan terhadap potensi manfaatnya.
</li>
</ul>
</li>
<li><b>Memilih untuk Terlibat</b></b>
Keputusan untuk mengintegrasikan AI generatif ke dalam pembelajaran bersifat pribadi, tetapi memerlukan perspektif yang seimbang dan terinformasi:
<ul align="justify">
<li><b>Kesadaran akan Bias:</b> Kenali bagaimana pengaruh pribadi dan sosial membentuk penolakan atau penerimaan Anda.
</li>
<li><b>Keterbukaan terhadap Eksplorasi:</b> Dekati AI dengan rasa ingin tahu, bukan rasa takut atau penilaian yang terbentuk sebelumnya.
</li>
<li><b>Pemikiran Kritis:</b> Terlibatlah dengan AI secara kritis, gunakan AI sebagai alat untuk meningkatkan-bukan menggantikan-penalaran manusia.
</li>
</ul>
</li>
</ol>

<p align="justify">
Di persimpangan antara AI dan pembelajaran, penolakan adalah hal yang wajar tetapi dapat diatasi. Dengan mengenali akar bias kita dan secara sadar memutuskan untuk terlibat dengan AI generatif, kita dapat membuka potensinya untuk membentuk kembali cara kita belajar, mengajar, dan berinteraksi dengan pengetahuan. <summary><button><b><mark>4.5. Deep Learning: Tiga Elemen Utama</mark></b></button></summary>
<p align="justify">
Deep learning dalam konteks pendidikan, dengan elemen <b>Mindful, Meaningful</b>, dan <b>Joyful Learning</b>, adalah pendekatan yang bertujuan menciptakan pengalaman belajar yang holistik dan transformatif. Pendekatan ini berakar dari konsep psikologi, filsafat pendidikan, dan teori pembelajaran modern. Inspirasi untuk pendekatan ini dapat ditelusuri dari beberapa tokoh dan teori berikut:
</p>
<h3> 1. <b>Mindful Learning</b></h3>
     <ul align="justify">
     <li><b>Inspirasi Utama</b>: Ellen Langer.<br>
     Ellen Langer, seorang psikolog dari Harvard, mempopulerkan konsep "<b>mindfulness</b>" dalam pembelajaran. Ia menekankan pentingnya kesadaran penuh (mindfulness) dalam proses belajar untuk menghadirkan perhatian pada saat ini, mengurangi pola pikir otomatis, dan meningkatkan keterlibatan serta kreativitas.
     </li>
     <li><b>Prinsip</b>:
     <ul align="justify">
     <li> Mendorong keterbukaan terhadap perspektif baru.</li>
     <li> Mengurangi pola pikir tetap (fixed mindset).</li>
     <li> Mengintegrasikan pembelajaran dengan pengalaman hidup sehari-hari.
     </li>
     </ul>
     </li>
    </ul>
<h3> 2. <b>Meaningful Learning</b></h3>
  <ul align="justify">  
  <li><b>Inspirasi Utama</b>: David Ausubel.<br>  
     Ausubel mengembangkan teori pembelajaran bermakna (meaningful learning), yang menekankan pentingnya menghubungkan informasi baru dengan pengetahuan yang sudah ada dalam pikiran peserta didik. Ia percaya bahwa pembelajaran akan lebih efektif jika relevan dan kontekstual.
   </li>
   <li><b>Prinsip</b>:
     <ul align="justify">
     <li> Pembelajaran harus relevan dengan pengalaman dan tujuan peserta didik.</li>
     <li> Materi harus dikaitkan dengan kehidupan nyata atau masalah yang mereka hadapi.</li>
     </ul>
  </li>
  </ul>
<h3> 3. <b>Joyful Learning</b></h3>
   <ul align="justify">  
    <li><b>Inspirasi Utama</b>: Maria Montessori & Lev Vygotsky<br>
    <ul aling="justify">  
     <li>Montessori percaya bahwa anak-anak belajar dengan lebih baik ketika mereka merasa senang, bebas bereksplorasi, dan berinteraksi dengan lingkungannya.</li>  
     <li> Vygotsky menekankan pentingnya zona perkembangan proksimal (ZPD) dan interaksi sosial dalam pembelajaran yang menyenangkan. Proses pembelajaran menjadi lebih efektif jika melibatkan dukungan sosial yang mendukung rasa percaya diri.</li>
     </ul>
     </li>
   <li><b>Prinsip</b>:
     <ul align="justify">
     <li> Lingkungan pembelajaran harus mendukung rasa eksplorasi, kreativitas, dan kegembiraan.</li>
     <li> Memberikan tantangan yang sesuai, sehingga tidak terlalu sulit atau terlalu mudah.</li>
     </ul>
    </li>
   </ul>
<h3> <b>Pendekatan Holistik</b></h3>
<p align="justify">
Elemen <b>Mindful, Meaningful</b>, dan <b>Joyful</b> sering kali diintegrasikan dalam teori pembelajaran berbasis konstruktivisme dan humanisme, yang memandang peserta didik sebagai individu dengan potensi unik. Tokoh seperti <b>John Dewey</b> (belajar melalui pengalaman) dan <b>Carl Rogers</b> (pendekatan humanistik) juga turut menginspirasi praktik pembelajaran ini.
</p>
<h3> <b>Dalam Konteks Teknologi</b></h3>
<p align="justify">
Pendekatan ini kini semakin relevan dalam pengembangan teknologi seperti <b>Artificial Intelligence in Education (AI-ED)</b>. Misalnya, desain platform pembelajaran seperti Suno AI atau aplikasi lain yang mengutamakan pengalaman pengguna yang adaptif, menarik, dan menyenangkan sering kali mengambil inspirasi dari prinsip-prinsip ini.
</p>
<hr>
<p align="justify">
Deep learning adalah pendekatan dalam pembelajaran mesin yang meniru cara kerja otak manusia untuk memproses data dan membuat keputusan. Dalam konteks pendidikan, deep learning tidak hanya berfokus pada penguasaan materi, tetapi juga pada pengalaman belajar yang lebih mendalam dan bermakna. Tiga elemen utama dalam deep learning adalah <b>mindful learning</b>, <b>meaningful learning</b>, dan <b>joyful learning</b>. Berikut adalah penjelasan rinci mengenai ketiga elemen tersebut.
</p>

<ol align="justify">
<li> <b>Mindful Learning</b><br>
Mindful learning adalah pendekatan yang menekankan kesadaran penuh dalam proses belajar. Dalam konteks ini, siswa diajak untuk hadir sepenuhnya dalam setiap pengalaman belajar, baik secara mental maupun emosional. Mindful learning mendorong siswa untuk:
<center>

image.png

</center>
<ul align="justify">
<li><b>Menyadari Proses Belajar:</b> Siswa diajarkan untuk memperhatikan bagaimana mereka belajar, termasuk strategi yang mereka gunakan dan bagaimana mereka merespons tantangan.
</li>
<li><b>Refleksi Diri:</b> Siswa didorong untuk melakukan refleksi terhadap pengalaman belajar mereka, sehingga mereka dapat memahami kekuatan dan kelemahan mereka.
</li>
<li><b>Mengurangi Stres:</b> Dengan berfokus pada saat ini, siswa dapat mengurangi kecemasan dan stres yang sering kali mengganggu proses belajar.
</li>
</ul>
Mindful learning membantu siswa untuk lebih terlibat dan berkomitmen dalam proses belajar, sehingga mereka dapat mencapai pemahaman yang lebih dalam.
<br>
<h4><b>Contoh Aplikasi:</b></h4>
<ul align="justify">
<li>Dalam kelas diskusi, siswa tidak hanya mendengar tetapi juga mengajukan pertanyaan kritis dan berkontribusi pada dialog.
</li>
<li>Belajar melalui meditasi atau teknik mindfulness untuk meningkatkan konsentrasi selama proses belajar.
</li>
</ul>
</li>
<li><b>Meaningful Learning</b><br>
Meaningful learning adalah konsep yang menekankan pentingnya mengaitkan informasi baru dengan pengetahuan yang sudah ada. Dalam pendekatan ini, siswa tidak hanya menghafal fakta, tetapi juga memahami konteks dan relevansi dari apa yang mereka pelajari. Beberapa aspek penting dari meaningful learning meliputi:
<center>

image.png

</center>
<ul align="justify">
<li><b>Koneksi dengan Pengetahuan Sebelumnya:</b> Siswa diajarkan untuk mengaitkan konsep baru dengan pengalaman dan pengetahuan yang sudah mereka miliki, sehingga
informasi baru menjadi lebih mudah dipahami dan diingat.
</li>
<li><b>Relevansi:</b> Pembelajaran yang bermakna terjadi ketika siswa melihat relevansi dari materi yang dipelajari dengan kehidupan sehari-hari mereka atau dengan minat pribadi mereka.
</li>
<li><b>Keterlibatan Aktif:</b> Siswa didorong untuk terlibat secara aktif dalam proses belajar, seperti berdiskusi, berkolaborasi, dan menerapkan pengetahuan dalam situasi nyata.
</li>
</ul>
Dengan meaningful learning, siswa dapat mengembangkan pemahaman yang lebih dalam dan bertahan lama terhadap materi yang dipelajari.

<h4><b>Contoh Aplikasi:</b></h4>
<ul align="justify">
<li>Pembelajaran berbasis proyek, di mana siswa bekerja pada masalah nyata yang memerlukan penerapan konsep-konsep yang dipelajari.
</li>
<li>Mengajarkan konsep matematika dengan mengaitkannya pada situasi dunia nyata, seperti menghitung anggaran rumah tangga.
</li>
</ul>
</li>

<li><b>Joyful Learning</b><br>
Joyful learning adalah pendekatan yang menekankan pentingnya kebahagiaan dan kesenangan dalam proses belajar. Ketika siswa merasa senang dan terinspirasi, mereka lebih cenderung untuk terlibat dan berprestasi. Beberapa elemen dari joyful learning meliputi:
<center>

image.png

</center>
<ul align="justify">
<li><b>Lingkungan Belajar yang Positif:</b> Menciptakan suasana belajar yang menyenangkan dan mendukung, di mana siswa merasa aman untuk bereksperimen dan berinovasi.
</li>
<li><b>Penggunaan Metode Kreatif:</b> Menggunakan berbagai metode pengajaran yang menarik, seperti permainan, proyek kreatif, dan teknologi, untuk membuat
pembelajaran lebih menyenangkan.
</li>
<li><b>Pujian dan Penghargaan:</b> Memberikan umpan balik positif dan penghargaan atas usaha siswa, sehingga mereka merasa dihargai dan termotivasi untuk terus belajar.
</li>
</ul>
Joyful learning membantu siswa untuk mengembangkan cinta terhadap proses belajar, yang pada gilirannya dapat meningkatkan motivasi dan hasil belajar mereka.

<h4><b>Contoh aplikasi:</b></h4>
<ul align="justify">
<li>Menggunakan gamifikasi dalam pembelajaran untuk membuat materi lebih menarik.
</li>
<li>Menciptakan kompetisi sehat dalam kelas dengan hadiah simbolis yang membuat siswa antusias.
</li>
</ul>
</li>
</ol>
<h3><b>Rangkuman</b></h3>
<p align="justify">
Deep learning yang melibatkan mindful learning, meaningful learning, dan joyful learning
menciptakan pengalaman belajar yang lebih holistik dan mendalam. Dengan mengintegrasikan ketiga elemen ini, siswa tidak hanya memperoleh pengetahuan, tetapi juga
keterampilan berpikir kritis, kreativitas, dan kecintaan terhadap belajar yang akan bermanfaat sepanjang hidup mereka.
</p>
<h4><b>Integrasi Ketiga Komponen</b></h4>
<p align="justify">Ketiga komponen ini saling melengkapi untuk menciptakan pengalaman belajar yang utuh:
</p>
<ol align="justify">
<li>Mindful Learning membantu siswa tetap fokus dan kritis.
</li>
<li>Meaningful Learning memastikan apa yang dipelajari relevan dan bermanfaat.
</li>
<li>Joyful Learning membuat proses belajar menjadi pengalaman yang menyenangkan dan menginspirasi.
</li>
</ol>
<p align="justify">
Ketiga komponen <b>Mindful Learning, Meaningful Learning</b>, dan <b>Joyful Learning</b> sangat relevan dengan <b>Taksonomi Bloom</b>, yang merupakan kerangka kerja untuk memahami dan mengklasifikasikan tujuan pembelajaran. Taksonomi Bloom membagi pembelajaran menjadi tiga domain utama: <b>kognitif, afektif</b>, dan <b>psikomotorik</b>, dengan setiap domain memiliki tahapan pengembangan tertentu. Berikut adalah kaitan masing-masing komponen dengan Taksonomi Bloom:
</p>
<ol align="justify">
<li><b>Mindful Learning dan Taksonomi Bloom</b><br>
Mindful Learning menekankan kesadaran, perhatian penuh, dan refleksi dalam pembelajaran, yang terutama berkaitan dengan domain kognitif dalam Taksonomi Bloom. Dalam domain ini, terdapat enam tingkat berpikir yang progresif:
<ol align="justify">
<li><b>Mengingat (Remembering):</b> Mengingat kembali informasi yang telah dipelajari.
</li>
<li><b>Memahami (Understanding):</b> Menjelaskan arti atau konsep dari materi.
</li>
<li><b>Menerapkan (Applying):</b> Menggunakan informasi dalam situasi baru.
</li>
<li><b>Menganalisis (Analyzing):</b> Membagi informasi menjadi komponen untuk memahami struktur.
</li>
<li><b>Mengevaluasi (Evaluating):</b> Membuat keputusan berdasarkan kriteria tertentu.
</li>
<li><b>Mencipta (Creating):</b> Menghasilkan sesuatu yang baru berdasarkan informasi yang dipelajari.
</li>
</ol>
<h5><b>Kaitan:</b></h5>
<ul align="justify">
<li>Mindful Learning membantu siswa mencapai tingkat berpikir yang lebih tinggi, seperti analisis, evaluasi, dan kreasi, dengan mendorong mereka untuk reflektif, kritis, dan sadar akan proses pembelajaran.
</li>
<li>Misalnya, siswa diajarkan untuk tidak hanya menghafal fakta, tetapi juga menganalisis relevansi dan menciptakan solusi inovatif.
</li>
</ul>
</li>
<li><b>Meaningful Learning dan Taksonomi Bloom</b><br>
Meaningful Learning berfokus pada menghubungkan informasi baru dengan pengetahuan yang sudah dimiliki, yang juga sesuai dengan domain kognitif Taksonomi Bloom, terutama di tingkat yang lebih tinggi.<br>
<br>
<b>Kaitan:</b>
<ul align="justify">
<li><b>Pemahaman (Understanding):</b> Informasi baru dikaitkan dengan pengalaman atau pengetahuan sebelumnya.
</li>
<li><b>Aplikasi (Applying):</b> Informasi yang bermakna lebih mudah diterapkan dalam situasi nyata.
</li>
<li><b>Kreasi (Creating):</b> Pengetahuan yang bermakna memungkinkan siswa menciptakan karya baru atau memecahkan masalah kompleks.
</li>
</ul>
Selain itu, Meaningful Learning juga berkaitan dengan <b>domain afektif</b>, yang mencakup <b>respons emosional terhadap pembelajaran</b>:
<ol align="justify">
<li><b>Menerima (Receiving):</b> Siswa menerima informasi baru secara terbuka.
</li>
<li><b>Menanggapi (Responding):</b> Siswa menunjukkan minat untuk mempelajari lebih lanjut.
</li>
<li><b>Menilai (Valuing):</b> Siswa melihat nilai dari apa yang dipelajari dan mengaplikasikannya.
</li>
</ol>
</li>
<li><b>Joyful Learning dan Taksonomi Bloom</b><br>
Joyful Learning berkaitan erat dengan <b>domain afektif</b> dalam Taksonomi Bloom. Domain ini mencakup aspek <b>emosional</b> seperti <b>motivasi, minat</b>, dan <b>apresiasi</b> terhadap pembelajaran.
<br>
<br>
<b>Kaitan:</b>
<ul align="justify">
<li><b>Motivasi intrinsik:</b> Joyful Learning membantu siswa merasa bahwa pembelajaran adalah pengalaman yang menyenangkan, sehingga mereka lebih cenderung menghargai dan melibatkan diri secara emosional (Valuing).
</li>
<li><b>Penguatan Afektif:</b> Siswa menikmati pembelajaran melalui kegiatan kreatif, interaktif, atau kolaboratif, seperti permainan, proyek, atau diskusi.
</li>
<li><b>Peningkatan Retensi:</b> Ketika siswa senang belajar, mereka lebih mungkin mengingat dan memahami informasi dengan lebih baik, mendukung domain kognitif tingkat rendah seperti <b>mengingat</b> dan <b>memahami</b>, hingga tingkat tinggi seperti <b>kreasi</b>.
</li>
</ul>
</li>
</ol>

<h3> <b>Teori dan Sifat-Sifat Pertidaksamaan</b><h3>

<h4> <b>Teori Pertidaksamaan</b><h4>
<p align="justify">
Pertidaksamaan adalah hubungan antara dua ekspresi matematis yang melibatkan simbol-simbol ketidaksamaan ($<,\ >,\ \leq,\ \geq$). Pertidaksamaan digunakan untuk membatasi atau menentukan daerah nilai-nilai variabel yang memenuhi suatu kondisi tertentu. Semua variabel yang dibahas adalah bilangan real.
</p>
<p align="justify">
Pertidaksamaan dapat digunakan dalam banyak aspek seperti:
</p>
<ul align="justify">
<li> <b>Analisis ekonomi:</b> Menentukan batas keuntungan atau kerugian.
</li>
<li> <b>Teknik:</b> Menghitung batas toleransi mekanis.</li>
<li> <b>Statistika:</b> Menentukan probabilitas kejadian dalam interval tertentu.
</li>
</ul>
<hr>

<h4> <b>Sifat-Sifat Pertidaksamaan</b></h4>
<p align="justify">
Pertidaksamaan memiliki beberapa sifat dasar:
</p>
<ol align="justify">
<li> <b>Sifat Transitif</b><br>
   Jika $ a > b $ dan $ b > c $, maka $ a > c $.
   <br>
   <b>Bukti:</b><br>
   Dari sifat urutan bilangan real, jika $ a > b $ berarti $ a - b > 0 $, dan jika $ b > c $, maka $ b - c > 0 $. Dengan menjumlahkan kedua persamaan, diperoleh $ a - c > 0 $, sehingga $ a > c $. 😊
</li>
<li> <b>Sifat Penjumlahan</b><br>
   Jika $ a > b $, maka $ a + c > b + c $ untuk sembarang $ c $.<br>
   <b>Bukti:</b><br>
   Karena $ a - b > 0 $, maka dengan menambahkan $ c $ pada kedua sisi diperoleh
   $$
   a + c - (b + c) = a - b > 0\implies a+c > b+c. 😊
   $$
</li>
<li> <b>Sifat Perkalian</b>
   <ul align="justify">
   <li> Jika $ a > b $ dan $ c > 0 $, maka $ ac > bc $.</li>
   <li> Jika $ a > b $ dan $ c < 0 $, maka $ ac < bc $.</li>
   </ul>
   <b>Bukti:</b><br>
   Untuk $ c > 0 $, kalikan kedua sisi $ a - b > 0 $ dengan $ c $, sehingga $ c(a - b) > 0 $ atau $ ac - bc > 0 $, yang berarti $ ac > bc $.  
   Untuk $ c < 0 $, hasil kali dengan bilangan negatif membalik tanda. 😊
</li>
<li> <b>Sifat Pembagian</b>
   <ul align="justify">
   <li> Jika $ a > b $ dan $ c > 0 $, maka $ \frac{a}{c} > \frac{b}{c} $.</li>
   <li> Jika $ a > b $ dan $ c < 0 $, maka $ \frac{a}{c} < \frac{b}{c} $. 😊</li>
   </ul>
</li>
<li> <b>Sifat Kuadrat</b>

   - Jika $ a > b $ dan $ a, b > 0 $, maka $ a^2 > b^2 $.
   - Jika $ |a| > |b| $, maka $ a^2 > b^2 $. 😊

</li>
</ol>
<hr>

<h4> <b>Contoh Dunia Nyata Pertidaksamaan</b></h4>
<ol align="justify">
<li> <b>Ekonomi: Menghitung Biaya Minimum</b>  
   Misalkan sebuah perusahaan memiliki anggaran pemasaran $ x $ juta rupiah, dan setiap produk memerlukan biaya minimal Rp5 juta per unit. Jika mereka memproduksi $ n $ produk, maka:  
   $$
   5n \leq x
   $$  
   Artinya, jumlah produk yang dapat dibuat adalah $ n \leq \frac{x}{5} $.
</li>
<li> <b>Fisik: Toleransi Mesin</b>  
   Dalam perancangan mesin, panjang suatu komponen harus berada di antara 10 cm hingga 12 cm untuk menjaga stabilitas.  
   $$
   10 \leq x \leq 12
   $$
</li>
<li> <b>Statistika: Probabilitas Suatu Interval</b>  
   Dalam distribusi normal, nilai yang berada dalam $ \mu - \sigma \leq x \leq \mu + \sigma $ meliputi sekitar 68% dari data.
</li>
</ol>
<hr>

<h4> <b>Bukti Praktis Sifat Pertidaksamaan</b></h4>
<p>
<b>Sifat Perkalian:</b>
</p>
<p align="justify">
Buktikan bahwa jika $ a > b $ dan $ c > 0 $, maka $ ac > bc $. Bukti,
$$
a > b \implies a - b > 0 \implies  (a - b)c > 0 \implies ac - bc > 0 \implies ac > bc.
$$
Misal $ a = 3 $, $ b = 2 $, dan $ c = 4 $:
$$
a > b \quad \Rightarrow \quad 3 > 2
$$  
Kalikan kedua sisi dengan $ c = 4 $:  
$$
ac = 3 \cdot 4 = 12, \quad bc = 2 \cdot 4 = 8
$$  
Hasil: $ 12 > 8 $, sifat terbukti. 😊
</p>
<hr>

<h4> <b>Contoh Menarik</b></h4>
<p align="justify">
<b>Pertidaksamaan Kuadrat dalam Dunia Nyata: Proyek Pembangunan</b>  
<br>
Seorang arsitek merancang jembatan dengan lengkungan parabola berbentuk $ y = -x^2 + 4x + 5 $. Untuk keamanan, tinggi maksimum jembatan harus berada di bawah 8 meter.  
Cari interval $ x $ di mana $ y < 8 $:  
$$
-x^2 + 4x + 5 < 8\implies -x^2 + 4x - 3 < 0
$$
Faktor:  
$$
-(x^2 - 4x + 3) < 0 \quad \Rightarrow \quad -(x - 1)(x - 3) < 0
$$
Interval solusi: $ x < 1$ atau $x > 3 $.  
Artinya, jembatan memiliki tinggi aman di antara $ x = 1 $ hingga $ x = 3 $ meter. 😊
</p>

<h4> <b>Penyelesaian Lengkap dan Diagram untuk Pertidaksamaan</b>:</h4>  
$$
-x^2 + 4x - 3 < 0
$$

<h4> <b>Langkah 1: Susun ulang agar memudahkan penyelesaian</b></h4>
<p align="justify">
Karena koefisien $ x^2 $ adalah negatif, kita kali kedua sisi dengan $-1$ agar lebih mudah dianalisis. Ingat bahwa tanda pertidaksamaan berubah arah saat dikalikan dengan bilangan negatif:
$$
x^2 - 4x + 3 > 0
$$
</p>

<h4> <b>Langkah 2: Faktorkan persamaan kuadrat</b></h4>
<p align="justify">
Faktorkan $ x^2 - 4x + 3 $:
$$
x^2 - 4x + 3 = (x - 1)(x - 3)
$$
Sehingga, pertidaksamaan menjadi:
$$
(x - 1)(x - 3) > 0
$$
</p>
<h4> <b>Langkah 3: Tentukan nol fungsi</b></h4>
<p align="justify">
Titik-titik di mana ekspresi $ (x - 1)(x - 3) = 0 $ adalah:
$$
x = 1 \quad \text{dan} \quad x = 3
$$
</p>
<h4> <b>Langkah 4: Uji tanda di setiap interval</b></h4>
<p align="justify">
Bagi garis bilangan berdasarkan titik kritis $ x = 1 $ dan $ x = 3 $ menjadi tiga interval:
</p>

1. $ (-\infty, 1) $
2. $ (1, 3) $
3. $ (3, \infty) $

<p align="justify">
Pilih nilai uji di setiap interval untuk menentukan tanda ekspresi $ (x - 1)(x - 3) $:
</p>
<ul align="justify">
<li> <b>Interval $ (-\infty, 1) $:</b> Pilih $ x = 0 $  
  $$
  (0 - 1)(0 - 3) = (-)(-) = +
  $$
  Tanda: $ + $
</li>
<li> <b>Interval $ (1, 3) $:</b> Pilih $ x = 2 $  
  $$
  (2 - 1)(2 - 3) = (+)(-) = -
  $$
  Tanda: $ - $
<li>
<li> <b>Interval $ (3, \infty) $:</b> Pilih $ x = 4 $  
  $$
  (4 - 1)(4 - 3) = (+)(+) = +
  $$
  Tanda: $ + $.
</li>
</ul>
<hr>

<h4> <b>Langkah 5: Tentukan solusi</b></h4>
<p align="justify">
Karena pertidaksamaan $ (x - 1)(x - 3) > 0 $, maka kita cari interval di mana ekspresi bernilai positif ($+$):
$$
x \in (-\infty, 1) \cup (3, \infty)
$$
</p>
<hr>

<h4> <b>Himpunan Penyelesaian</b></h4>
$$
\boxed{\pmb{x \in (-\infty, 1) \cup (3, \infty)}}
$$
<hr>

<h4> <b>Diagram Garis Bilangan</b></h4>
<ol align="justify">
<li> Gambarkan garis bilangan</li>
<li> Beri tanda lingkaran kosong di $ x = 1 $ dan $ x = 3 $
 karena nilai $ x = 1 $ dan $ x = 3 $ tidak termasuk dalam solusi (hanya $ > 0 $).
<li> Arsir daerah $ (-\infty, 1) $ dan $ (3, \infty) $. 😊
</ol>

<p align="justify">
Berikut adalah program untuk menampilkan gambar grafik $y = -x^2 + 4x + 5$ dan penyelesaian dari pertidaksamaan
$$
-x^2 + 4x - 3 < 0.
$$
</p>
<hr>

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk grafik pertama
x_vals2 = np.linspace(-1, 5, 500)
y_vals2 = -x_vals2**2 + 4 * x_vals2 + 5

# Fungsi untuk grafik kedua (pertaksamaan parabola)
x_vals1 = np.linspace(-1, 5, 500)
y_vals1 = -x_vals1**2 + 4 * x_vals1 - 3
root_vals1 = [1, 3]  # Akar-akar parabola pertama

# Plot
plt.figure(figsize=(7,3))

# Garis sumbu x
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')

# Grafik kedua
plt.plot(x_vals1, y_vals1, label=r"$-x^2 + 4x - 3$", color='blue')
plt.scatter(root_vals1, [0, 0], color='red', zorder=5, label="Akar ($x = 1, x = 3$)")
plt.fill_between(
    x_vals1, y_vals1, 0,
    where=((x_vals1 < root_vals1[0]) | (x_vals1 > root_vals1[1])),
    color='gray', alpha=0.3, label="Daerah solusi ($y < 0$)"
)

# Grafik pertama
plt.plot(x_vals2, y_vals2, label=r"$y=-x^2 + 4x + 5$", color='green', linewidth=4)

# Memberikan anotasi dan label
plt.title("Diagram Parabola dan Solusi Pertidaksamaan", fontsize=13)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.annotate("y < 0", xy=(-0.7, -1.5), fontsize=10, color='black', weight='bold')
plt.annotate("y < 0", xy=(3.95, -1.5), fontsize=10, color='black', weight='bold')
plt.legend(fontsize=10)
plt.grid(alpha=0.3)

# Menampilkan plot
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, clear_output
import ipywidgets as widgets

def explain_program():
    """Menampilkan penjelasan tentang program ini."""
    explanation = (
        """
        <style>
            .code-box {
                background-color: #f0f0f0;
                padding: 2px 4px;
                border-radius: 3px;
                border: 1px solid #ccc;
                font-family: monospace;
            }
        </style>
        <div style="font-family:Arial, sans-serif; line-height:1.6;">
            <h2><mark>Deskripsi Program</mark></h2>
            <p>
                Program ini adalah alat interaktif untuk menyelesaikan <b><mark>persamaan non-linear</mark></b> dengan beberapa fitur menarik:
            </p>
            <ol>
                <li>
                    <strong>Input Persamaan:</strong><br>
                    Pengguna dapat memasukkan persamaan matematika dalam bentuk string (contoh: <code class="code-box">x<b>6 - 4*x<b>2 + 3</code>). Program secara otomatis mengubah string tersebut menjadi ekspresi matematika yang dapat diproses.
                </li>
                <li>
                    <strong>Penyelesaian Simbolik dan Numerik:</strong><br>
                    Program akan menghitung himpunan solusi simbolik dari persamaan dan juga menampilkan nilai numerik dari akar-akar real.
                </li>
                <li>
                    <strong>Visualisasi Grafik:</strong><br>
                    Selain solusi, program akan menampilkan grafik fungsi untuk memberikan ilustrasi visual tentang letak akar-akar real pada sumbu x.
                </li>
                <li>
                    <strong>Antarmuka Interaktif:</strong><br>
                    Dengan menggunakan <em><b>ipywidgets</b></em>, pengguna dapat dengan mudah:
                    <ul>
                        <li>Memasukkan persamaan menggunakan kotak teks.</li>
                        <li>Mengklik tombol untuk menghitung solusi.</li>
                        <li>Melihat hasil dalam area output secara langsung.</li>
                    </ul>
                </li>
            </ol>
            <h3>Komponen Utama</h3>
            <ul>
                <li><strong>Sympy:</strong> Untuk manipulasi simbolik dan pencarian solusi.</li>
                <li><strong>Numpy:</strong> Untuk menghitung nilai numerik.</li>
                <li><strong>Matplotlib:</strong> Untuk membuat grafik fungsi.</li>
                <li><strong>IPyWidgets:</strong> Untuk membuat antarmuka interaktif.</li>
            </ul>
            <h3>Cara Penggunaan</h3>
            <ol>
                <li>Masukkan persamaan dalam format Python di kolom input, seperti <code class="code-box">x<b>3 - 2*x + 1</code>.</li>
                <li>Klik tombol <strong>Selesaikan</strong>.</li>
                <li>Hasil akan ditampilkan di bawah, meliputi:
                    <ul>
                        <li>Persamaan asli.</li>
                        <li>Solusi simbolik dan numerik.</li>
                        <li>Grafik fungsi lengkap dengan penanda akar real (jika ada).</li>
                    </ul>
                </li>
            </ol>
        </div>
        """
    )
    display(widgets.HTML(value=explanation))

def solve_equation(equation_str):
    """Mencari akar persamaan non-linier, menampilkan grafik, dan nilai numerik akar real.

    Args:
      equation_str: String yang merepresentasikan persamaan non-linier.

    Returns:
      None. Menampilkan hasil dan grafik secara langsung.
    """

    # Tentukan variabel
    x = sp.Symbol('x')

    # Bersihkan output sebelumnya
    clear_output(wait=True)

    # Ubah string persamaan menjadi ekspresi sympy
    try:
        equation = sp.sympify(equation_str)
    except:
        print("Input persamaan tidak valid.")
        return

    # Selesaikan persamaan
    solutions = sp.solve(equation, x)

    # Tampilkan persamaan dan solusi
    display(Math(r"\text{Persamaan:}\ " + sp.latex(equation) + " = 0"))
    display(Math(r"\text{Himpunan Penyelesaian real:}\ " + sp.latex(sp.FiniteSet(*solutions))))

    # Filter untuk akar real saja
    real_roots = [r for r in solutions if r.is_real]

    # Tampilkan nilai numerik akar real
    if real_roots:
        display(Math(r"\text{Himpunan Nilai numerik akar-akar real:}\ " + sp.latex(sp.FiniteSet(*[r.evalf() for r in real_roots]))))

        # Buat fungsi numerik dari persamaan
        func = sp.lambdify(x, equation)

        # Tentukan rentang x untuk grafik
        x_range = np.linspace(-5, 5, 50)  # Sesuaikan rentang jika diperlukan

        # Hitung nilai y
        y_values = func(x_range)

        # Gambar grafik
        plt.figure(figsize=(6, 4))  # Ukuran grafik yang lebih proporsional
        plt.plot(x_range, y_values, label=f"$f(x) = {sp.latex(equation)}$")
        plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.title(f"Grafik $f(x) = {sp.latex(equation)}$")
        plt.grid(True)

        # Tandai akar real pada grafik
        for root in real_roots:
            plt.plot(float(root), 0, 'ro', label=f"Akar-real: {root.evalf()}")

        plt.legend()
        plt.show()
    else:
        print("Persamaan tidak memiliki akar real.")
        display(Math(r"\text{Himpunan Nilai numerik akar-akar:}\ " + sp.latex(sp.FiniteSet(*[r.evalf() for r in solutions]))))

# Input interaktif menggunakan ipywidgets
equation_input = widgets.Text(
    value="x<b>6 - 4*x<b>2 + 3",
    placeholder='Masukkan persamaan (misalnya x<b>6 - 4*x<b>2 + 3 = 0)',
    description='Persamaan:',
    disabled=False
)

solve_button = widgets.Button(description="Selesaikan")
explain_button = widgets.Button(description="Penjelasan Program!")
output_area = widgets.Output()

# Fungsi untuk tombol penyelesaian
def on_solve_button_clicked(b):
    with output_area:
        clear_output(wait=True)
        solve_equation(equation_input.value)

# Fungsi untuk tombol penjelasan
def on_explain_button_clicked(b):
    with output_area:
        clear_output(wait=True)
        explain_program()

solve_button.on_click(on_solve_button_clicked)
explain_button.on_click(on_explain_button_clicked)

display(widgets.VBox([equation_input, widgets.HBox([solve_button, explain_button]), output_area]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
from ipywidgets import VBox, HBox, Layout
import re

# Deskripsi program
deskripsi_program = """
<div style="font-family: Arial, sans-serif; line-height: 1.6; margin: 10px 0;">
    <h2 style="color: #4CAF50;">Program Simulasi Pertumbuhan Populasi Bakteri</h2>
    <p>Program ini dirancang untuk memvisualisasikan dan menghitung pertumbuhan populasi bakteri
    berdasarkan data yang Anda masukkan. Anda dapat:</p>
    <ul>
        <li>Memasukkan data waktu (jam) dan jumlah populasi bakteri.</li>
        <li>Melihat visualisasi data awal dalam bentuk grafik.</li>
        <li>Menghitung fungsi eksponensial yang menggambarkan pertumbuhan populasi.</li>
        <li>Melihat prediksi populasi di waktu tertentu.</li>
    </ul>
    <p>Langkah-langkah penggunaan:</p>
    <ol>
        <li>Klik tombol <button><b>Tampilkan Deskripsi</b></button> untuk melihat/tidak-melihat Deskripsi Program.</li>
        <li>Masukkan data waktu (jam) pada kolom <button><b>Jam</b></button> (contoh: <i>0,1,2,3,4,5</i>).</li>
        <li>Masukkan data populasi pada kolom <button><b>Populasi</b></button> (contoh: <i>100,200,400,800,1600,3200</i>).</li>
        <li>Pilih waktu <button><b>Prediksi</b></button> (opsional) untuk menghitung nilai prediksi ke depan.</li>
        <li>Klik tombol <button><b>Hitung</b></button> untuk melihat hasil perhitungan dan menampilkan visualisasi simulasi.</li>
    </ol>
    <p><mark><b>Selamat mencoba!</b></mark></p>
</div>
"""

# Fungsi untuk menyimpan grafik
def simpan_grafik(fig, filename):
    fig.savefig(filename, dpi=300, bbox_inches='tight')
    with output_area:
        print(f"Grafik berhasil disimpan sebagai '{filename}'")

# Fungsi untuk menghitung pertumbuhan populasi
def hitung_pertumbuhan(jam_input, populasi_input, waktu_prediksi):
    output_area.clear_output(wait=True)  # Bersihkan output sebelumnya

    try:
        # Validasi format input menggunakan regex
        if not re.fullmatch(r"(\d+(\.\d+)?(,\d+(\.\d+)?)*,?)", jam_input) or \
           not re.fullmatch(r"(\d+(\.\d+)?(,\d+(\.\d+)?)*,?)", populasi_input):
            raise ValueError("Input harus berupa angka yang dipisahkan dengan koma.")

        # Konversi input ke array
        jam = np.array(list(map(float, jam_input.split(","))))
        populasi = np.array(list(map(float, populasi_input.split(","))))
    except ValueError as e:
        with output_area:
            print(f"Kesalahan: {e}")
        return

    if len(jam) != len(populasi):
        with output_area:
            print("Kesalahan: Jumlah data jam dan populasi harus sama.")
        return

    if len(jam) < 2:
        with output_area:
            print("Kesalahan: Masukkan minimal dua data untuk perhitungan.")
        return

    if any(populasi <= 0):
        with output_area:
            print("Kesalahan: Data populasi harus bernilai positif.")
        return

    with output_area:
        # Tampilkan grafik awal
        fig1, ax1 = plt.subplots(figsize=(6, 4))
        ax1.plot(jam, populasi, 'o', label='Data Awal')
        ax1.set_xlabel('Jam')
        ax1.set_ylabel('Populasi Bakteri')
        ax1.set_title('Pertumbuhan Populasi Bakteri')
        ax1.grid(True)
        plt.show()

        # Hitung parameter fungsi eksponensial
        koefisien = np.polyfit(jam, np.log(populasi), 1)
        r = koefisien[0]  # Laju pertumbuhan
        P0 = np.exp(koefisien[1])  # Populasi awal

        # Tampilkan fungsi pertumbuhan
        print(f"Fungsi pertumbuhan: P(t) = {P0:.2f} * e^({r:.2f}t)")

        # Prediksi populasi di waktu tertentu
        if waktu_prediksi.strip():
            try:
                waktu_prediksi = float(waktu_prediksi)
                prediksi = P0 * np.exp(r * waktu_prediksi)
                print(f"Prediksi populasi di jam ke-{waktu_prediksi}: {prediksi:.2f}")
            except ValueError:
                print("Prediksi tidak valid. Masukkan angka untuk waktu prediksi.")
        else:
            print("Waktu prediksi tidak dimasukkan.")

        # Tampilkan grafik fungsi pertumbuhan
        fig2, ax2 = plt.subplots(figsize=(6, 4))
        t_fit = np.linspace(min(jam), max(jam) + 10, 500)  # Perluas cakupan grafik hingga 10 jam di atas data
        ax2.plot(jam, populasi, 'o', label='Data Awal')
        ax2.plot(t_fit, P0 * np.exp(r * t_fit), '-', label='Fungsi Pertumbuhan')
        ax2.set_xlabel('Jam')
        ax2.set_ylabel('Populasi Bakteri')
        ax2.set_title('Pertumbuhan Populasi Bakteri')
        ax2.legend()
        ax2.grid(True)
        plt.show()

        # Simpan grafik (opsional)
        simpan_grafik(fig2, 'grafik_pertumbuhan.png')

# Fungsi untuk menangani klik tombol "Hitung"
def on_hitung_button_clicked(b):
    hitung_pertumbuhan(jam_input.value, populasi_input.value, prediksi_input.value)

# Fungsi untuk menampilkan atau menyembunyikan deskripsi
def toggle_deskripsi(change):
    if change["new"]:  # Jika tombol aktif, tampilkan deskripsi
        deskripsi_area.value = deskripsi_program
    else:  # Jika tombol tidak aktif, sembunyikan deskripsi
        deskripsi_area.value = ""

# Widget input
jam_input = widgets.Text(description="Jam:", placeholder="Contoh: 0,1,2,3,4,5", layout=Layout(width='70%'))
populasi_input = widgets.Text(description="Populasi:", placeholder="Contoh: 100,200,400,800,1600,3200", layout=Layout(width='70%'))
prediksi_input = widgets.Text(description="Prediksi:", placeholder="Jam prediksi, contoh: 6", layout=Layout(width='30ch'))
hitung_button = widgets.Button(description="Hitung", layout=Layout(width='50%', min_width='50px', max_width='75px'))
hitung_button.on_click(on_hitung_button_clicked)

# Output area untuk menampilkan hasil
output_area = widgets.Output()

# Widget untuk deskripsi
deskripsi_area = widgets.HTML(value="")
toggle_button = widgets.ToggleButton(value=False, description="Tampilkan Deskripsi")
toggle_button.observe(toggle_deskripsi, names="value")

# Layout dengan VBox
input_section = VBox(
    [
        HBox([toggle_button, deskripsi_area]),
        jam_input,
        populasi_input,
        prediksi_input,
        hitung_button,
    ],
    layout=Layout(padding="10px", border="solid 1px #ccc", width="90%", margin="auto"),
)

# Layout untuk keseluruhan
main_layout = VBox([input_section, output_area], layout=Layout(width="100%", align_items="center"))

# Menampilkan widget
display(main_layout)
```

<p align="justify">
Dalam pelemparan dua dadu, terdapat $36$ kemungkinan hasil ($6$ sisi pada dadu pertama $\times$ $6$ sisi pada dadu kedua).
</p>
<p align="justify">
Untuk mendapatkan jumlah $7$, terdapat $6$ kemungkinan kombinasi:
$$(1, 6), (2, 5), (3, 4), (4, 3), (5, 2),(6, 1).$$
</p>
<p align="justify">
Secara matematis, probabilitas mendapatkan jumlah $7$ adalah $6$ (jumlah kombinasi yang diinginkan) dibagi dengan $36$ (total kemungkinan kombinasi), yaitu: $6/36=1/6$.
</p>
<p align="justify">
Simulasi yang dilakukan dengan program Python mendekati nilai probabilitas teoritis ini. Semakin banyak percobaan yang dilakukan (misalnya 1000000 kali), hasil simulasi akan semakin mendekati nilai probabilitas matematis.
</p>
<p align="justify">
<b>Secara Statistika:</b>
Simulasi pelemparan dadu merupakan contoh dari pendekatan frekuensi relatif dalam statistika. Frekuensi relatif adalah jumlah kejadian suatu peristiwa dibagi dengan total jumlah percobaan.
</p>
<p align="justify">
Dalam simulasi ini, frekuensi relatif mendapatkan jumlah $7$ akan mendekati probabilitas teoritis $1/6$ seiring bertambahnya jumlah percobaan. Hal ini menunjukkan hukum bilangan besar, yaitu semakin besar ukuran sampel, semakin dekat frekuensi relatif dengan probabilitas sebenarnya.
</p>
<p align="justify">
Rumus yang digunakan untuk menghitung frekuensi relatif adalah:
<br>
<b>Frekuensi Relatif = (Jumlah Kejadian Suatu Peristiwa) / (Total Jumlah Percobaan)</b>
<br>
Dalam konteks simulasi pelemparan dadu:
</p>
<ul align="justify">
<li>
<b>Jumlah Kejadian Suatu Peristiwa:</b> Jumlah percobaan di mana jumlah kedua dadu adalah:$2\leq m\leq 12$.
</li>
<li>
<b>Total Jumlah Percobaan:</b> Total jumlah simulasi pelemparan dadu (misalnya $n=1000000$ kali).
</li>
</ul>
<p align="justify">
Frekuensi relatif yang diperoleh dari simulasi akan mendekati probabilitas teoritis suatu peristiwa seiring dengan bertambahnya jumlah percobaan.
</p>

</p>

```python
import random
import pandas as pd
from fractions import Fraction
import ipywidgets as widgets
from IPython.display import display, clear_output

# Deskripsi program dalam format HTML
deskripsi_program = """
<h4 style="color: #4CAF50;">Simulasi Probabilitas Jumlah Dadu</h4>
<p>
Program ini mensimulasikan pelemparan dua buah dadu sebanyak <b>n</b> kali dan menghitung:
</p>
<ul>
    <li>Hasil pelemparan untuk setiap pasangan dadu.</li>
    <li>Banyaknya percobaan yang menghasilkan jumlah tertentu (<b>m</b>).</li>
    <li>Probabilitas eksperimen dan probabilitas eksak untuk jumlah <b>m</b>.</li>
</ul>
<p>
Langkah-langkah penggunaan:
</p>
<ol>
    <li>Masukkan banyaknya percobaan (<b>n</b>) dan jumlah total (<b>m</b>).</li>
    <li>Klik tombol <b>Jalankan Simulasi</b> untuk melihat hasilnya.</li>
    <li>Hasil simulasi termasuk data percobaan, jumlah percobaan dengan hasil <b>m</b>,
    dan perbandingan probabilitas eksperimen dengan probabilitas eksak.</li>
</ol>
<p>Selamat mencoba!</p>
"""

# Fungsi untuk menghitung probabilitas eksak
def hitung_probabilitas_eksak(m):
    """
    Menghitung probabilitas eksak jumlah dua dadu = m.
    Probabilitas dihitung sebagai jumlah kombinasi yang menghasilkan m
    dibagi dengan total ruang sampel (36).
    """
    if m < 2 or m > 12:
        return Fraction(0)  # Nilai tidak mungkin
    # Ada (m-1) kombinasi valid untuk 2 <= m <= 7 dan (13-m) untuk 8 <= m <= 12
    kombinasi = m - 1 if m <= 7 else 13 - m
    return Fraction(kombinasi, 36)

# Fungsi untuk menjalankan simulasi
def simulasi(n, m):
    hasil = []
    jumlah_m = 0

    for _ in range(n):
        dadu1 = random.randint(1, 6)
        dadu2 = random.randint(1, 6)
        jumlah = dadu1 + dadu2
        hasil.append([dadu1, dadu2, jumlah])
        if jumlah == m:
            jumlah_m += 1

    df = pd.DataFrame(hasil, columns=["Dadu 1", "Dadu 2", "Jumlah"])

    # Probabilitas eksperimen
    probabilitas_eksperimen = jumlah_m / n

    # Probabilitas eksak
    probabilitas_eksak = hitung_probabilitas_eksak(m)

    return df, jumlah_m, probabilitas_eksperimen, probabilitas_eksak

# Fungsi untuk menangani tombol klik
def on_jalankan_clicked(b):
    output_area.clear_output(wait=True)
    try:
        n = int(jumlah_percobaan_input.value)
        m = int(total_input.value)

        if n <= 0:
            raise ValueError("Jumlah percobaan harus lebih dari 0.")
        if m < 2 or m > 12:
            raise ValueError("Jumlah m harus di antara 2 dan 12.")

        df, jumlah_m, prob_eksperimen, prob_eksak = simulasi(n, m)

        with output_area:
            print(f"Banyaknya percobaan: {n}")
            print(f"Total m: {m}")
            print(f"\nProbabilitas eksperimen: {prob_eksperimen:.4f}")
            print(f"Probabilitas eksak: {prob_eksak} ({float(prob_eksak):.4f})")
            print(f"\nBanyaknya percobaan dengan jumlah {m}: {jumlah_m}")
            print(f"Banyaknya percobaan dengan jumlah selain {m}: {n - jumlah_m}")
            print("\n40 hasil terakhir:")
            display(df.tail(40))

    except ValueError as e:
        with output_area:
            print(f"Kesalahan: {str(e)}")

# Fungsi untuk menampilkan deskripsi
def toggle_deskripsi(change):
    if change["new"]:
        deskripsi_area.value = deskripsi_program
    else:
        deskripsi_area.value = ""

# Widget input
jumlah_percobaan_input = widgets.Text(
    description="Banyaknya Percobaan:",
    placeholder="Masukkan n (contoh: 1000)",
    layout=widgets.Layout(width="50%"),
)
total_input = widgets.Text(
    description="Jumlah (m):",
    placeholder="Masukkan m (contoh: 7)",
    layout=widgets.Layout(width="30%"),
)
jalankan_button = widgets.Button(
    description="Jalankan Simulasi",
    button_style="success",
    layout=widgets.Layout(width="30%"),
)
jalankan_button.on_click(on_jalankan_clicked)

# Output area
output_area = widgets.Output()

# Widget deskripsi
deskripsi_area = widgets.HTML(value="")
toggle_button = widgets.ToggleButton(
    value=False, description="Tampilkan Deskripsi", button_style="info"
)
toggle_button.observe(toggle_deskripsi, names="value")

# Layout
input_section = widgets.VBox(
    [toggle_button, deskripsi_area, jumlah_percobaan_input, total_input, jalankan_button],
    layout=widgets.Layout(padding="10px", border="solid 1px #ccc", width="90%"),
)
main_layout = widgets.VBox([input_section, output_area])

# Menampilkan aplikasi
display(main_layout)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, widgets

def kubus(sisi):
  # Membuat titik-titik kubus
  points = np.array([
      [0, 0, 0],
      [sisi, 0, 0],
      [sisi, sisi, 0],
      [0, sisi, 0],
      [0, 0, sisi],
      [sisi, 0, sisi],
      [sisi, sisi, sisi],
      [0, sisi, sisi]
  ])

  # Menghubungkan titik-titik untuk membentuk kubus
  edges = [
      [0, 1], [1, 2], [2, 3], [3, 0],
      [4, 5], [5, 6], [6, 7], [7, 4],
      [0, 4], [1, 5], [2, 6], [3, 7]
  ]

  # Membuat figure dan axes
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')

  # Menggambar kubus
  for edge in edges:
    ax.plot(points[edge, 0], points[edge, 1], points[edge, 2], 'b-')

  # Setting tampilan
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_zlabel('Z')
  ax.set_title('Kubus')

  # Menghitung volume kubus
  volume = sisi**3

  # Menampilkan informasi kubus
  print(f"Panjang sisi: {sisi}")
  print(f"Rumus volume: sisi * sisi * sisi")
  print(f"Volume kubus: {volume}")

  # Menampilkan kubus
  plt.show()

# Membuat slider interaktif untuk input sisi
interact(kubus, sisi=widgets.IntSlider(min=1, max=10, step=1, value=1))
```

```python
import numpy as np
import plotly.graph_objects as go

# Fungsi untuk membuat bola
def buat_bola():
    u = np.linspace(0, 2 * np.pi, 50)
    v = np.linspace(0, np.pi, 50)
    x = np.outer(np.cos(u), np.sin(v))
    y = np.outer(np.sin(u), np.sin(v))
    z = np.outer(np.ones(np.size(u)), np.cos(v))
    return go.Surface(x=x, y=y, z=z, colorscale='Viridis')

# Fungsi untuk membuat kubus
def buat_kubus():
    x = [0, 0, 1, 1, 0, 0, 1, 1]
    y = [0, 1, 1, 0, 0, 1, 1, 0]
    z = [0, 0, 0, 0, 1, 1, 1, 1]
    i = [0, 0, 4, 0, 1, 2, 5, 3, 6, 4, 7, 5]
    j = [1, 4, 5, 3, 2, 6, 7, 7, 3, 5, 1, 0]
    k = [3, 5, 6, 2, 7, 4, 0, 6, 4, 7, 2, 1]
    return go.Mesh3d(x=x, y=y, z=z, i=i, j=j, k=k, color='blue', opacity=0.5)

# Fungsi untuk membuat balok
def buat_balok():
    x = [0, 0, 2, 2, 0, 0, 2, 2]
    y = [0, 1, 1, 0, 0, 1, 1, 0]
    z = [0, 0, 0, 0, 1, 1, 1, 1]
    i = [0, 0, 4, 0, 1, 2, 5, 3, 6, 4, 7, 5]
    j = [1, 4, 5, 3, 2, 6, 7, 7, 3, 5, 1, 0]
    k = [3, 5, 6, 2, 7, 4, 0, 6, 4, 7, 2, 1]
    return go.Mesh3d(x=x, y=y, z=z, i=i, j=j, k=k, color='green', opacity=0.5)

# Fungsi untuk membuat tabung
def buat_tabung():
    theta = np.linspace(0, 2 * np.pi, 50)
    z = np.linspace(0, 3, 20)
    theta, z = np.meshgrid(theta, z)
    x = np.cos(theta)
    y = np.sin(theta)
    return go.Surface(x=x, y=y, z=z, colorscale='Blues')

# Fungsi untuk membuat prisma segitiga
def buat_prisma():
    x = [0, 1, 0.5, 0, 1, 0.5]
    y = [0, 0, np.sqrt(3) / 2, 0, 0, np.sqrt(3) / 2]
    z = [0, 0, 0, 1, 1, 1]
    i = [0, 1, 2, 0, 3, 4]
    j = [1, 2, 0, 4, 4, 5]
    k = [2, 0, 1, 5, 3, 3]
    return go.Mesh3d(x=x, y=y, z=z, i=i, j=j, k=k, color='purple', opacity=0.5)

# Fungsi untuk menampilkan geometri
def tampilkan_geometri():
    print("Pilih bentuk geometri:")
    print("1. Bola")
    print("2. Kubus")
    print("3. Balok")
    print("4. Tabung")
    print("5. Prisma")

    pilihan = input("Masukkan nomor pilihan Anda (1-5): ")

    if pilihan == '1':
        fig = go.Figure(data=[buat_bola()])
    elif pilihan == '2':
        fig = go.Figure(data=[buat_kubus()])
    elif pilihan == '3':
        fig = go.Figure(data=[buat_balok()])
    elif pilihan == '4':
        fig = go.Figure(data=[buat_tabung()])
    elif pilihan == '5':
        fig = go.Figure(data=[buat_prisma()])
    else:
        print("Pilihan tidak valid!")
        return

    fig.update_layout(scene=dict(aspectmode='data'))
    fig.show()

# Menjalankan program
tampilkan_geometri()
```

### <b>1. Pengertian Homomorfisma Grup</b>

Dalam teori grup, <b>homomorfisma</b> adalah fungsi antara dua grup yang mempertahankan struktur grup, artinya operasi grup pada domain (grup asal) akan terjaga pada kodomain (grup tujuan).

Jika kita memiliki dua grup, $G$ dan $H$, dengan operasi grup yang masing-masing dilambangkan oleh $*$ dan $\circ$, maka fungsi $\phi: G \rightarrow H$ disebut homomorfisma grup jika untuk setiap elemen $a, b\in G$, berlaku:
$$
\phi(a * b) = \phi(a) \circ \phi(b)
$$

Sederhananya, homomorfisma memastikan bahwa hasil operasi pada elemen di grup $G$ dipetakan menjadi hasil operasi di grup $H$ sesuai dengan peta yang dibuat oleh $\phi$.

---

### <b>2. Sifat-Sifat Penting dalam Homomorfisma Grup</b>

Homomorfisma memiliki beberapa sifat yang penting dan sering digunakan dalam pembuktian matematika:

1. <b>Peta Elemen Identitas</b>  
   Jika $e_G$ adalah elemen identitas di $G$ dan $e_H$ adalah elemen identitas di $H$, maka homomorfisma $\phi: G \rightarrow H$ akan memetakan identitas $G$ ke identitas $H$, yaitu:
   $$
   \phi(e_G) = e_H
   $$

2. <b>Pemetaan Invers</b>  
   Untuk setiap elemen $a \in G$, kita punya:
   $$
   \phi(a^{-1}) = \phi(a)^{-1}
   $$

3. <b>Kernel Homomorfisma</b>  
   Kernel dari homomorfisma $\phi: G \rightarrow H$ adalah himpunan semua elemen di $G$ yang dipetakan ke elemen identitas di $H$:
   $$
   \text{Ker}(\phi) = \{ g \in G \mid \phi(g) = e_H \}
   $$
   Kernel merupakan subgrup normal dari $G$.

4. <b>Image Homomorfisma</b>  
   Image dari homomorfisma $\phi$ adalah himpunan semua elemen di $H$ yang merupakan hasil pemetaan dari elemen-elemen di $G$:
   $$
   \text{Im}(\phi) = \{ \phi(g) \mid g \in G \}
   $$
   Image dari homomorfisma juga merupakan subgrup dari $H$.

---

### <b>3. Teorema Penting Terkait Homomorfisma</b>

Ada beberapa teorema utama yang sering digunakan dalam teori homomorfisma grup:

- <b>Teorema Homomorfisma Pertama</b>  
  Teorema ini menyatakan bahwa jika $\phi: G \rightarrow H$ adalah homomorfisma grup dengan kernel $\text{Ker}(\phi)$, maka $G / \text{Ker}(\phi) \cong \text{Im}(\phi)$. Artinya, grup hasil bagi $G$ terhadap kernel $\text{Ker}(\phi)$ isomorfik dengan Image dari $\phi$.

- <b>Teorema Homomorfisma Kedua</b>  
  Jika $N$ dan $K$ adalah subgrup dari $G$ dengan $N \subset K$ dan $N \triangleleft G$, maka:
  $$
  K / N \cong KN / N
  $$

- <b>Teorema Homomorfisma Ketiga</b>  
  Jika $N$ dan $K$ adalah subgrup normal dari $G$ dengan $N \subset K$, maka:
  $$
  (G / N) / (K / N) \cong G / K
  $$

Teorema-teorema ini sangat penting dalam mengklasifikasikan grup dan memahami struktur grup melalui homomorfisma.

### <b>4. Contoh Kongkrit Homomorfisma Grup</b>

1. <b>Homomorfisma dari Grup Bilangan Bulat ke Bilangan Modulo</b>  
   Misalkan $G = \mathbb{Z}$ adalah grup bilangan bulat dengan operasi penjumlahan dan $H = \mathbb{Z}_n$ adalah grup bilangan bulat modulo $n$. Definisikan fungsi $\phi: \mathbb{Z} \rightarrow \mathbb{Z}_n$ sebagai:
   $$
   \phi(a) = a \mod n
   $$
   Fungsi ini adalah homomorfisma, karena:
   $$
   \phi(a + b) = (a + b) \mod n = (a \mod n + b \mod n) \mod n = \phi(a) + \phi(b)
   $$

2. <b>Homomorfisma dari Grup Matriks ke Determinan</b>  
   Misalkan $G$ adalah grup matriks $2 \times 2$ invertibel dengan determinan bukan nol, dan $H = \mathbb{R}^*$ adalah grup bilangan real bukan nol. Definisikan fungsi $\phi: G \rightarrow H$ sebagai determinan dari matriks:
   $$
   \phi(A) = \det(A)
   $$
   Ini adalah homomorfisma, karena determinan dari perkalian dua matriks adalah perkalian dari determinan masing-masing:
   $$
   \phi(AB) = \det(AB) = \det(A) \cdot \det(B) = \phi(A) \cdot \phi(B)
   $$

### <b>Sifat 1: Peta Elemen Identitas</b>

<b>Bukti:</b>
Karena $ \phi $ adalah homomorfisma, maka untuk setiap $ a \in G $:

$$
\phi(e_G * a) = \phi(e_G) \circ \phi(a)
$$

Di sisi lain, karena $ e_G $ adalah elemen identitas di $ G $, maka $ e_G * a = a $. Jadi, kita punya:

$$
\phi(a) = \phi(e_G) \circ \phi(a)
$$

Agar persamaan ini benar, maka $ \phi(e_G) $ harus bertindak sebagai elemen identitas di $ H $, yang berarti:

$$
\phi(e_G) = e_H
$$

Dengan demikian, elemen identitas di $ G $ dipetakan ke elemen identitas di $ H $ oleh homomorfisma $ \phi $.

### <b>Sifat 2: Pemetaan Invers</b>

Jika $ \phi: G \rightarrow H $ adalah homomorfisma, maka untuk setiap $ a \in G $,

$$
\phi(a^{-1}) = \phi(a)^{-1}
$$

<b>Bukti:</b>
Karena $ a^{-1} $ adalah invers dari $ a $ dalam grup $ G $, kita tahu bahwa:

$$
a * a^{-1} = e_G
$$
Dengan menerapkan homomorfisma $ \phi $, kita dapatkan:

$$
\phi(a * a^{-1}) = \phi(e_G)
$$

Dengan sifat homomorfisma, kita bisa menuliskan:

$$
\phi(a) \circ \phi(a^{-1}) = e_H
$$

Artinya, $ \phi(a^{-1}) $ adalah invers dari $ \phi(a) $ dalam grup $ H $, sehingga:

$$
\phi(a^{-1}) = \phi(a)^{-1}
$$

Ini membuktikan bahwa homomorfisma memetakan invers di $ G $ menjadi invers di $ H $.

### <b>Sifat 3: Kernel Homomorfisma</b>

Kernel dari homomorfisma $ \phi: G \rightarrow H $ adalah himpunan semua elemen di $G$ yang dipetakan ke elemen identitas di $H$:

$$
\text{Ker}(\phi) = \{ g \in G \mid \phi(g) = e_H \}
$$

Kita akan membuktikan bahwa kernel dari homomorfisma $\phi$ adalah subgrup normal dari $ G $.

<b>Bukti:</b>
1. <b>Kernel adalah subgrup dari $ G $:</b>
   - Elemen identitas $ e_G $ ada di kernel karena, berdasarkan sifat 1, $\phi(e_G) = e_H $. Jadi, $ e_G \in \text{Ker}(\phi)$.
   - Untuk setiap $ a,b \in \text{Ker}(\phi)$, kita memiliki $\phi(a) = e_H$ dan $\phi(b) = e_H$.

     $$
     \phi(a * b^{-1}) = \phi(a) \circ \phi(b)^{-1} = e_H \circ e_H = e_H
     $$

     Karena $ a * b^{-1} \in \text{Ker}(\phi)$, maka $\text{Ker}(\phi)$ adalah subgrup dari $G$.

2. <b>Kernel adalah subgrup normal dari $ G $:</b>
   Untuk setiap $ g \in G $ dan $ k \in \text{Ker}(\phi) $, kita punya:

   $$
   \phi(g * k * g^{-1}) = \phi(g) \circ \phi(k) \circ \phi(g^{-1}) = \phi(g) \circ e_H \circ \phi(g)^{-1} = \phi(g) \circ \phi(g)^{-1} = e_H
   $$

   Jadi, $ g * k * g^{-1} \in \text{Ker}(\phi) $ untuk setiap $ g \in G $ dan $ k \in \text{Ker}(\phi) $, yang menunjukkan bahwa $ \text{Ker}(\phi) $ adalah subgrup normal dari $ G $.

### <b>Sifat 4: Image Homomorfisma</b>

Image dari homomorfisma $ \phi: G \rightarrow H $, yaitu himpunan semua elemen di $ H $ yang merupakan hasil pemetaan dari elemen-elemen di $ G $, adalah subgrup dari $ H $:

$$
\text{Im}(\phi) = \{ \phi(g) \mid g \in G \}
$$

<b>Bukti:</b>
1. <b>Identitas di $ \text{Im}(\phi) $:</b>  
   Dari sifat 1, kita tahu bahwa $ \phi(e_G) = e_H $, sehingga $ e_H \in \text{Im}(\phi) $.

2. <b>Tertutup terhadap operasi grup di $ H $:</b>  
   Untuk setiap $ a, b \in \text{Im}(\phi) $, terdapat elemen $ g, h \in G $ sehingga $ a = \phi(g) $ dan $ b = \phi(h) $. Karena $ \phi $ adalah homomorfisma, kita dapatkan:

   $$
   a \circ b = \phi(g) \circ \phi(h) = \phi(g * h)
   $$

   Karena $ g * h \in G $, maka $ \phi(g * h) \in \text{Im}(\phi) $. Jadi, $ \text{Im}(\phi) $ tertutup terhadap operasi grup di $ H $.

3. <b>Tertutup terhadap invers di $ H $:</b>  
   Untuk setiap $ a \in \text{Im}(\phi) $, terdapat $ g \in G $ sehingga $ a = \phi(g) $. Berdasarkan sifat 2, kita tahu:

   $$
   a^{-1} = \phi(g)^{-1} = \phi(g^{-1})
   $$
   
   Karena $ g^{-1} \in G $, maka $ \phi(g^{-1}) \in \text{Im}(\phi) $, sehingga $ a^{-1} \in \text{Im}(\phi) $.

Karena identitas, operasi, dan invers dalam $ \text{Im}(\phi) $ memenuhi semua syarat subgrup, maka $ \text{Im}(\phi) $ adalah subgrup dari $ H $.

\begin{equation}
ax^2+bx+c=0, a\neq 0,

sekaligus menampilkan grafiknya.

```python
from sympy import symbols, sqrt, latex, simplify
from IPython.display import display, Math, clear_output
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np

# Definisi variabel simbolik
x = symbols('x')  # Variabel kuadrat
a, b, c = symbols('a b c')  # Koefisien

# Kotak input dengan ukuran kecil menggunakan widgets
a_input = widgets.IntText(description='a:', value=1, layout=widgets.Layout(width='150px'))
b_input = widgets.IntText(description='b:', value=1, layout=widgets.Layout(width='150px'))
c_input = widgets.IntText(description='c:', value=1, layout=widgets.Layout(width='150px'))

# Keterangan dalam kotak
keterangan_box = widgets.HTML(
    """
    <div style="
        border: 1px solid #4CAF50;
        padding: 10px;
        border-radius: 5px;
        background-color: #f9f9f9;
        color: #333;
        font-size: 14px;
        line-height: 1.5;
    ">
        <b>Keterangan Program:</b><br>
        Program ini digunakan untuk menyelesaikan persamaan kuadrat dengan format:
        <b>ax² + bx + c = 0</b>.<br>
        <ul>
            <li>Masukkan nilai dalam box <b>a</b>, <b>b</b>, dan <b>c</b> dengan menggerakkan kursor, klik tanda ke <b>atas</b>/<b>bawah</b>!.</li>
            <li>Pastikan <b>a ≠ 0</b>.</li>
        </ul>
        Program akan menghitung <b>diskriminan</b>, menampilkan jenis akar, serta menggambar grafik persamaan kuadrat.
        <p>
        <b>Copyright ©2027 by the author Subiono</b>
        </p>
    </div>
    """
)

# Fungsi untuk menggambar grafik
def plot_persamaan(a, b, c):
    x_vals = np.linspace(-10, 10, 500)
    y_vals = a * x_vals**2 + b * x_vals + c  # Perbaikan: Menggunakan x_vals**2

    fig, ax = plt.subplots(figsize=(4, 3))  # Ukuran grafik lebih kecil
    ax.plot(x_vals, y_vals, label=f"$y={a}x^2 + {b}x + {c}$", color='blue')
    ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
    ax.axvline(0, color='black', linewidth=0.8, linestyle='--')
    ax.grid(color='gray', linestyle='--', linewidth=0.5)
    ax.set_title("Grafik Persamaan Kuadrat", fontsize=10)
    ax.set_xlabel("x", fontsize=8)
    ax.set_ylabel("y", fontsize=8)
    ax.legend(loc="upper center", fontsize=8)
    plt.tight_layout()
    return fig

# Fungsi untuk menghitung dan menampilkan output
def hitung_persamaan(change=None):
    clear_output(wait=True)  # Membersihkan output sebelumnya
    display(keterangan_box)  # Menampilkan keterangan kembali
    display(widgets.HBox([a_input, b_input, c_input]))  # Menampilkan input dalam HBox

    a_val, b_val, c_val = a_input.value, b_input.value, c_input.value

    # Validasi nilai a ≠ 0
    if a_val == 0:
        display(Math(r"\text{Error: Nilai } a \text{ tidak boleh } 0."))
        return

    # Persamaan kuadrat dengan nilai yang dimasukkan
    persamaan = a_val * pow(x, 2) + b_val * x + c_val  # Perbaikan: Menggunakan pow(x, 2)
    output_text = widgets.Output()
    with output_text:
        display(Math(r"\text{Persamaan kuadrat: } " + latex(persamaan) + " = 0"))

        # Menghitung diskriminan
        D = b_val**2 - 4 * a_val * c_val
        display(Math(r"\text{Diskriminan: } \Delta = b^2-4ac = " + latex(D)))

        # Menentukan jenis akar berdasarkan diskriminan
        if D > 0:
            display(Math(r"\text{Akar-akar persamaan adalah real dan berbeda:}"))
            x1 = simplify((-b_val + sqrt(D)) / (2 * a_val))
            x2 = simplify((-b_val - sqrt(D)) / (2 * a_val))
            display(Math(r"x_1 = " + latex(x1)))
            display(Math(r"x_2 = " + latex(x2)))
        elif D == 0:
            display(Math(r"\text{Akar-akar persamaan adalah real dan sama:}"))
            x1 = simplify(-b_val / (2 * a_val))
            display(Math(r"x = " + latex(x1)))
        else:
            display(Math(r"\text{Akar-akar persamaan adalah tidak real (imajiner):}"))
            x1 = simplify((-b_val + sqrt(D)) / (2 * a_val))
            x2 = simplify((-b_val - sqrt(D)) / (2 * a_val))
            display(Math(r"x_1 = " + latex(x1)))
            display(Math(r"x_2 = " + latex(x2)))

    # Menampilkan grafik persamaan
    output_graph = widgets.Output()
    with output_graph:
        fig = plot_persamaan(a_val, b_val, c_val)
        plt.show()

    # Menyusun output dalam layout: Grafik di kanan, input di kiri
    display(widgets.HBox([output_text, output_graph]))

# Menambahkan event listener untuk setiap perubahan nilai input
a_input.observe(hitung_persamaan, names='value')
b_input.observe(hitung_persamaan, names='value')
c_input.observe(hitung_persamaan, names='value')

# Menampilkan keterangan dan input pertama kali
display(keterangan_box)
display(widgets.HBox([a_input, b_input, c_input]))  # Menampilkan input dalam HBox
hitung_persamaan()  # Memanggil fungsi untuk menghasilkan output awal
```
